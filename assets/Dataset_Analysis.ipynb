```python
dataset, ds_info = tfds.load('stanford_dogs', split='all', shuffle_files=True, data_dir='tensorflow_datasets/', as_supervised=True, with_info=True)
```

```python
ds_info
```

```python
fig = tfds.show_examples(dataset, ds_info)
```

```python
vals = np.unique(np.fromiter(dataset.map(lambda x, y: y), float), return_counts=True)

for val, count in zip(*vals):
    print(int(val), count)
```

```python
classes, counts = vals

plt.figure(figsize=(15, 8))
plt.bar(classes, counts, color='skyblue')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Number of Images per Class')
plt.xticks(classes, rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```
