```python
import tensorflow as tf
import tensorflow_datasets as tfds
import os
from PIL import Image
import io

(ds_test, ds_val, ds_train), ds_info = tfds.load(
    'stanford_dogs',
    split=["test[0%:50%]", "test[50%:]", "train"],
    shuffle_files=True,
    data_dir='tensorflow_datasets/',
    as_supervised=True,
    with_info=True
)

def save_random_images(dataset, info, num_samples=20, output_folder='images_test'):
    # Create directory if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Shuffle and take a number of samples from the dataset
    for i, (image, label) in enumerate(dataset.shuffle(10000).take(num_samples)):
        # Decode label
        label_name = info.features['label'].int2str(label)

        # Convert the image to a PIL Image
        image = tf.image.encode_jpeg(image).numpy()  # Encode as JPEG
        image = Image.open(io.BytesIO(image))

        # Define file name
        filename = f"Test_img{i+1}_{label_name}.jpg"
        file_path = os.path.join(output_folder, filename)

        # Save the image
        image.save(file_path)

save_random_images(ds_test, ds_info)

```
